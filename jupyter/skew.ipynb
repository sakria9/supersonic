```python
rate = 48000
seconds = 1.6

test_freqs = np.linspace(300, 3000, 20)
# test_freqs = [400]

freq_ratios_map = {}

for freq in test_freqs:
    cycle_time = 1 / freq
    cycle_samples = int(cycle_time * rate)
    n_cycles = int(seconds / cycle_time)
    t = np.linspace(0, n_cycles * cycle_time, n_cycles * cycle_samples, endpoint=False)
    y = sine_wave(freq)(t)

    record_wave, play_wave, rate = play_and_record(y, rate, discard=False)
    record_wave = record_wave[rate//2:rate//2+rate]

    freqs = np.fft.rfftfreq(len(record_wave), 1 / rate)
    fft_data = np.fft.rfft(record_wave)

    max_freq = freqs[np.argmax(np.abs(fft_data))]

    # print(freq / max_freq)
    print(f"freq: {freq}, max_freq: {max_freq}, ratio: {freq / max_freq}")
    freq_ratios_map[freq] = freq / max_freq

# plt.plot(record_wave)
# plot_fft(record_wave, rate)

x = sorted(list(freq_ratios_map.keys()))
y = np.array([freq_ratios_map[f] for f in x])
plt.plot(x, y)
```
